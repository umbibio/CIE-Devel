```python
import numpy as np
import scipy.stats as st
import pandas as pd
from gbnet.models import ORNORModel
from gbnet.aux import genData
```

```python
NX, NActvX, NY = 10, 3, 200
Xgt, DEG, rels = genData(NX, NActvX, NY, AvgNTF=0.5)
print(len(rels), 'edges in rels')
```

```python
model = ORNORModel(rels, DEG)
```

```python
while not model.converged():
    model.sample(N=500, burn=0.5, njobs=2)
```

```python
model.gelman_rubin[model.gelman_rubin>1.05]
```

```python
Res = model.result(Xgt=Xgt)
```
